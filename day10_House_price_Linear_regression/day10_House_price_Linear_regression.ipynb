```python
drop_col = [
    'PoolQC',
    'MiscFeature',
    'Alley',
    'Fence'
]
```

```python
df = df.drop(columns=drop_col)
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df['MasVnrType'] = df['MasVnrType'].fillna('None')
```

```python
df['FireplaceQu'] = df['FireplaceQu'].fillna('No_fireplace')
```

```python
garage_cols = [
    'GarageCond',
    'GarageFinish',
    'GarageType',
    'GarageQual'   
]

for col in garage_cols:
    df[col] = df[col].fillna('No_garage')
```

```python
bsmt_cols = [
    'BsmtFinType2',
    'BsmtExposure',
    'BsmtQual',
    'BsmtFinType1',
    'BsmtCond'
]

for col in bsmt_cols:
    df[col] = df[col].fillna('No_Basment')
```

```python
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())
```

```python
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].median())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head(1)
```

```python
X = df.drop(['SalePrice','Log_SalePrice'], axis=1)
```

```python
y = df['Log_SalePrice']
```

```python
print('Shape of X: ', X.shape)
print('Shape of y: ', y.shape)
```

```python
X_train,X_test, y_train, y_test = train_test_split(
    X, y, test_size=.20, random_state=42
)

print('X_train', X_train.shape)
print('y_train', y_train.shape)
print('X_test', X_test.shape)
print('y_test', y_test.shape)
```

```python
df.info()
```

```python
numeric_features = X.select_dtypes(
    include=['int64','float64']
).columns

categorical_features = X.select_dtypes(
    include=['str']
).columns

print("Shape of numeric_features: ", numeric_features.shape)
print("Shape of categorical_features: ", categorical_features.shape)

```

```python
numeric_features
```

```python
categorical_features
```

```python
print(len(numeric_features))
print(len(categorical_features))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
```

```python
numeric_pp = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pp = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown="ignore"))
])
```

```python
prepro = ColumnTransformer([
    ('num', numeric_pp, numeric_features),
    ('cat', categorical_pp, categorical_features )
])
```

```python
LinRpp = Pipeline([
    ('preprocessor', prepro),
    ('model', LinearRegression())
])
```

```python
LinRpp.fit(X_train, y_train)
```

```python
LinR_pred = LinRpp.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
mae = mean_absolute_error(y_test,LinR_pred)
mse = mean_squared_error(y_test,LinR_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, LinR_pred)

print('mean_absolute_error',':',mae) 
print('mean_squared_error',':',mse) 
print('rmse',':',rmse) 
print('r2_score',':',r2) 
```
