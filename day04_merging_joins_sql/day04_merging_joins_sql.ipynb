```python
import pandas as pd
```

```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "customer_name": ["John", "Mary", "David", "Sarah"]
})
```

```python
customers
```

```python
orders = pd.DataFrame({
    "customer_id": [1, 2, 1, 3],
    "order_amount": [500, 700, 300, 900]
})
```

```python
orders
```

```python
merged_df = pd.merge(
    customers,
    orders,
    on="customer_id"
)

```

```python
merged_df
```

```python
left_join_df = pd.merge(
    customers,
    orders,
    on = "customer_id",
    how = "left"
)
```

```python
left_join_df
```

```python
right_join_df = pd.merge(
    customers,
    orders,
    on = "customer_id",
    how = "right"
)
```

```python
right_join_df
```

```python
outer_join_df = pd.merge(
    customers,
    orders,
    on = "customer_id",
    how = "outer"
)
```

```python
outer_join_df
```

```python
orders_1 = pd.DataFrame({
    "customer_id": [1,2],
    "order_amount" : [500, 700] 
})

orders_2 = pd.DataFrame({
    "customer_id": [3, 4],
    "order_amount" : [900, 1200]
})
```

```python
combined_orders = pd.concat(
    [orders_1,orders_2]
)

combined_orders
```
