```python
pd.get_dummies(df['segment'])

```

```python
segments_encoded = pd.get_dummies(df['segment'], prefix='segment')

```

```python
df = pd.concat([df, segments_encoded], axis=1)
```

```python
df['sales_category'] = pd.cut(df['sales'], bins=3, labels=['Low', 'Medium', 'High'])
```

```python
df[['sales', 'sales_category']].head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df['sales_scaled'] = scaler.fit_transform(df[['sales']]) 
```

```python
df[['sales', 'sales_scaled', 'sales_category']].head(20)
```

```python
df['order_day'] = df['order_date'].dt.day
df['order_month'] = df['order_date'].dt.month
df['order_year'] = df['order_date'].dt.year
```

```python
df[['order_date', 'order_day', 'order_month', 'order_year']].head(20)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

```

```python
sales_pipeline = Pipeline([
    ('scaler', MinMaxScaler())
])
```

```python
df['sales_pipeline_scaled'] = sales_pipeline.fit_transform(df[['sales']])
```
